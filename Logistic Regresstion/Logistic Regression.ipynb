# Red Wine Quality Dataset

## Description:



### Part 1: Preprocessing

```python
import pandas as pd
import numpy as np 

dataset = pd.read_csv('winequality-red.csv')
```

```python
dataset
```

### Data Cleaning

```python
dataset = dataset.drop_duplicates()
dataset = dataset.dropna()


dataset
```

```python
dataset['quality'] = np.where(dataset['quality'] <= 5, 0, dataset['quality'])
dataset['quality'] = np.where(dataset['quality'] > 5, 1, dataset['quality'])

dataset
```

### Getting the Inputs and Outputs of the System

```python
X = dataset.iloc[:, :-1].values

X
```

```python
Y = dataset.iloc[:, -1].values

Y
```

### Creating Training and Testing Set

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
X_train
```

```python
X_test
```

```python
Y_test
```

```python
Y_train
```

### Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler

sc = MinMaxScaler()

X_train = sc.fit_transform(X_train)
```

### Building and Training the Model

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
Y_predict = model.predict(sc.transform(X_test))
```

```python
Y_predict
```

### Visualization

```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

### Evaluation

```python
from sklearn.metrics import confusion_matrix


confusion = confusion_matrix(Y_test, Y_predict)

print(confusion)

sns.heatmap(pd.DataFrame(confusion), annot=True, fmt='g')
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(Y_test, Y_predict)
```
