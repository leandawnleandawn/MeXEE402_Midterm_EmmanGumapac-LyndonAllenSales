# Advertising Dataset

## Description:

This dataset contains data on the amount spent on advertising through different channels (TV, radio, newspaper) and the resulting sales.1 It's a simple yet effective dataset for linear regression. 


## Identification of the Variables:

### Independent Variables:

1. TV - Expenditure of the TV Advertising
2. Radio - Expenditure of the Radio Advertising
3. Newspaper - Expenditure of the Newspaper Advertising

### Dependent Variable:

1. Sales - Overall Sales of the following data


By this we can use the Linear Regression for  this manner

#### Step 1: Preproccessing 

Importing the dataset using pandas


```python
import pandas as pd

dataset = pd.read_csv('advertising.csv')

dataset.head(10)
```

#### Identification of the Independent and Dependent Variables

```python
X = dataset.iloc[:, :-1].values

X
```

```python
Y = dataset.iloc[:, -1].values

Y
```

#### Creating the Training Set and the Test Set

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

### Part 2 - Building the Linear Regression Model

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
Y_predict = model.predict(X_test)

Y_predict 
```

### Part 3: Evaluation

```python
from sklearn.metrics import r2_score

r2 = r2_score(Y_test,Y_predict)

r2
```
